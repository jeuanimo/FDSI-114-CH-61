```python
numbers = [1,2,3,4,5]
placeholder = []

for num in numbers:
    placeholder.append(num)

reversed_list = []

while len(placeholder) > 0:
    item = placeholder.pop(0)
    reversed_list.insert(0,item)    

    print(reversed_list)
```
