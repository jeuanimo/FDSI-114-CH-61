## Find the totals of numbers from a collection of numbers that divided by 2 the module id 0

'''
Examples:
1. Given[3,4.10.15.99,2] Then the output should be **total = 3**
2. Given[2,4,6,10,3,7] Then the output should be: **total = 4**
### Criteria
1. Create a function that receives a collection and return an integer with total

### Hint:
 '%' is the arithmetic operator to get the module from a division
 2%2 = 0


```python
def divide_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += 1
    return total
my_list =[3,4,10,15,99,2,18]                                                
result = divide_numbers(my_list)
print(f"total equals: {result}")
```
