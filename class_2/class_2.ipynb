```python
# Import the random module to select random words
import random


# Secret words list (fruit words) - the pool of possible words for the game
WORDS = ['apple', 'banana', 'mango', 'orange', 'grape']
# Maximum number of incorrect guesses allowed before game over
MAX_LIVES = 6


def choose_word() -> str:
    """
    Randomly select a word from the WORDS list.
    
    Returns:
        str: A randomly chosen word from the WORDS list
    """
    # Use random.choice to pick one word from the WORDS list
    return random.choice(WORDS)


def display_state(secret: str, guessed: set) -> str:
    """
    Create a visual representation of the word with guessed letters revealed.
    
    Args:
        secret: The secret word to guess
        guessed: Set of letters that have been correctly guessed
        
    Returns:
        str: String showing guessed letters and underscores for unguessed letters
    """
    # For each character in secret, show it if guessed, otherwise show underscore
    # Join all characters with spaces for readability
    return ' '.join([ch if ch in guessed else '_' for ch in secret])


def valid_guess(s: str) -> bool:
    """
    Validate that the input is exactly one lowercase letter.
    
    Args:
        s: The string to validate
        
    Returns:
        bool: True if input is valid (single lowercase letter), False otherwise
    """
    # Check: must be exactly 1 character, must be alphabetic, must be lowercase
    return len(s) == 1 and s.isalpha() and s == s.lower()


def show_rules():
    """
    Display the game rules to the player.
    """
    # Print formatted rules section
    print("\n" + "="*50)
    print("HANGMAN RULES")
    print("="*50)
    print("1. The computer will choose a random fruit word")
    print("2. You have 6 lives to guess the word")
    print("3. Guess one lowercase letter at a time")
    print("4. If the letter is in the word, it will be revealed")
    print("5. If the letter is NOT in the word, you lose a life")
    print("6. Win by guessing all letters before running out of lives")
    print("7. Lose if you run out of lives before guessing the word")
    print("="*50 + "\n")


def display_game_status(secret: str, guessed: set, wrong_guesses: set, lives: int):
    """
    Display current game state to the player.
    
    Args:
        secret: The secret word being guessed
        guessed: Set of correctly guessed letters
        wrong_guesses: Set of incorrectly guessed letters
        lives: Number of lives remaining
    """
    # Print blank line for spacing
    print("\n" + display_state(secret, guessed))
    
    # Combine correct and wrong guesses, sort them, and display
    # Show 'none' if no letters have been guessed yet
    print(f"Guessed letters: {', '.join(sorted(guessed | wrong_guesses)) if (guessed or wrong_guesses) else 'none'}")
    
    # Display remaining lives
    print(f"Lives left: {lives}")


def get_valid_letter(guessed: set, wrong_guesses: set) -> str:
    """
    Get a valid letter guess from the player.
    Keeps prompting until a valid, non-duplicate letter is entered.
    
    Args:
        guessed: Set of correctly guessed letters
        wrong_guesses: Set of incorrectly guessed letters
        
    Returns:
        str: A valid, non-duplicate lowercase letter
    """
    # Loop until we get a valid input
    while True:
        # Get input from user and remove whitespace
        raw = input("Guess one lowercase letter: ").strip()
        
        # Check if the input is a valid single lowercase letter
        if not valid_guess(raw):
            print("Please enter exactly one lowercase letter (a-z).")
            continue  # Ask again
            
        # Check if this letter was already guessed (correct or wrong)
        if raw in guessed or raw in wrong_guesses:
            print("You already guessed that letter.")
            continue  # Ask again
            
        # Valid and unique letter - return it
        return raw


def process_guess(letter: str, secret: str, guessed: set, wrong_guesses: set, lives: int) -> int:
    """
    Process the player's guess and return updated lives count.
    
    Args:
        letter: The guessed letter
        secret: The secret word
        guessed: Set of correctly guessed letters
        wrong_guesses: Set of incorrectly guessed letters
        lives: Current number of lives
        
    Returns:
        int: Updated number of lives (decremented if guess was wrong)
    """
    # Check if the guessed letter is in the secret word
    if letter in secret:
        # Correct guess - add to guessed set
        guessed.add(letter)
        print(f"Good: '{letter}' is in the word.")
        return lives  # Lives unchanged
    else:
        # Wrong guess - add to wrong_guesses set
        wrong_guesses.add(letter)
        # Decrease lives by 1
        lives -= 1
        print(f"Sorry: '{letter}' is NOT in the word. Lives left: {lives}")
        return lives  # Return decremented lives


def check_win(secret: str, guessed: set) -> bool:
    """
    Check if the player has won the game by guessing all letters.
    
    Args:
        secret: The secret word
        guessed: Set of correctly guessed letters
        
    Returns:
        bool: True if all letters in secret have been guessed, False otherwise
    """
    # Check if every character in secret is in the guessed set
    return all(ch in guessed for ch in secret)


def play_interactive():
    """
    Main game loop for playing Hangman interactively.
    Manages the game state and coordinates the game flow.
    """
    # Select a random word from the word list
    secret = choose_word()
    
    # Initialize empty set for correctly guessed letters
    guessed: set = set()
    
    # Initialize empty set for incorrectly guessed letters
    wrong_guesses: set = set()
    
    # Start with maximum lives
    lives = MAX_LIVES

    # Display welcome message
    print("Welcome to Hangman!")
    print(f"The secret word has {len(secret)} letters.")

    # Main game loop - continues until win or lose
    while True:
        # Show current state of the game
        display_game_status(secret, guessed, wrong_guesses, lives)
        
        # Get a valid letter from the player
        letter = get_valid_letter(guessed, wrong_guesses)
        
        # Process the guess and update lives
        lives = process_guess(letter, secret, guessed, wrong_guesses, lives)

        # Check if player has won (all letters guessed)
        if check_win(secret, guessed):
            # Display final state with complete word
            print("\n" + display_state(secret, guessed))
            print(f"Congratulations, you guessed the word: {secret}")
            break  # Exit game loop

        # Check if player has lost (no lives remaining)
        if lives <= 0:
            print(f"\nGame over. The word was: {secret}")
            break  # Exit game loop


def main_menu():
    """
    Display and handle the main menu navigation.
    Allows players to start game, view rules, or quit.
    """
    # Loop until user chooses to quit
    while True:
        # Display menu header
        print("\n" + "="*50)
        print("HANGMAN GAME - MAIN MENU")
        print("="*50)
        
        # Display menu options
        print("1. Play Game")
        print("2. View Rules")
        print("3. Quit")
        print("="*50)
        
        # Get user's menu choice and remove whitespace
        choice = input("Enter your choice (1-3): ").strip()
        
        # Handle menu selection
        if choice == '1':
            # Start a new game
            play_interactive()
            print("\nReturning to main menu...")
        elif choice == '2':
            # Display the game rules
            show_rules()
        elif choice == '3':
            # Exit the program
            print("\nThanks for playing! Goodbye!")
            break  # Exit menu loop
        else:
            # Invalid input - ask again
            print("\nInvalid choice. Please enter 1, 2, or 3.")


# Run the game by starting the main menu
main_menu()
```
